```python
# missing data 처리

raw_df = raw_df.dropna()

raw_df.isnull().sum()
```

```python
# 정규화 (Date 제외한 모든 수치부분 정규화)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scale_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close',
              '3MA', '5MA', 'Volume'] # 정규화 대상 column 정의

scaled_df = scaler.fit_transform(raw_df[scale_cols]) # 정규화 수행

scaled_df = pd.DataFrame(scaled_df, columns=scale_cols) # 정규화된 새로운 DataFrame 생성

print(scaled_df)
```

```python
# 딥러닝 학습을 위한 입력데이터 feature column, 정답데이터 label column 정의 후 numpy로 변환하여 데이터 전처리 과정을 완료함
feature_cols = ['3MA', '5MA', 'Adj Close'] # feature 정의(입력데이터)
label_cols = ['Adj Close'] # label 정의(정답데이터)

label_df = pd.DataFrame(scaled_df, columns=label_cols)
feature_df = pd.DataFrame(scaled_df, columns=feature_cols)

print(feature_df)
print(label_df)

# 딥러닝 학습을 위해 DataFrame => numpy 변환
label_np = label_df.to_numpy()
feature_np = feature_df.to_numpy()
```

**----데이터 전처리 끝----**

```python
# 입력 파라미터 feature, label => numpy type

def make_sequene_dataset(feature, label, window_size):
    feature_list=[] # 생성될 feature list
    label_list = [] # 생성될 label list
    
    for i in range(len(feature)-window_size):
        # 슬라이싱을 이용하여 [[...],[...],...] 형상으로 입력데이터, 즉 feature를 생성함
        feature_list.append(feature[i:i+window_size])
        label_list.append(label[i+window_size])
        
    # feature_list=[...],[...],...] 이므로 리턴 값 np.array(feature_list)는 (batch size, time steps, imput dims) 형상 가짐
    return np.array(feature_list), np.array(label_list)
```

```python
window_size = 40

X, Y = make_sequene_dataset(feature_np, label_np, window_size)

print(X.shape, Y.shape)
```

```python
# 트레이닝 데이터 / 테스트 데이터 분리
split = -200

x_train = X[0:split]
y_train = Y[0:split]

x_test = X[split:]
y_test = Y[split:]

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
model = Sequential()

# LSTM 계층에 tanh를 활성화 함수로 가지는 노드 수 128개
# input_shape = (40, 3)
model.add(LSTM(128, activation='tanh', input_shape=x_train[0].shape))

model.add(Dense(1, activation='linear'))

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model.compile(loss='mse', optimizer='adam', metrics=['mae'])

early_stop = EarlyStopping(monitor='val_loss', patience=5)

model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=16, callbacks=[early_stop])
```

```python
# 삼성전자 주가 예측
pred = model.predict(x_test)

plt.figure(figsize=(12, 6))
plt.title('3MA + 5MA + Adj Close, window_size=40')
plt.ylabel('adj close')
plt.xlabel('period')
plt.plot(y_test, label='actual')
plt.plot(pred, label='prediction')
plt.grid()
plt.legend(loc='best')

plt.show()
```
